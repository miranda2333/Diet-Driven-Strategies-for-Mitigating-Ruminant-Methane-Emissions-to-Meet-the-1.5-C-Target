```python
import pandas as pd
import xgboost as xgb
import pickle
df_country = pd.read_csv('/content/Dairy_Grass_diet.csv', encoding='gbk')
df_country = df_country.iloc[:, 1:]
model_path = '/content/drive/MyDrive/Modeling/dairy_XGB_model.json'
loaded_model = xgb.Booster()
loaded_model.load_model(model_path)
with open('/content/drive/MyDrive/Modeling/dairy_ADD_proprocessor.pkl', 'rb') as fp:
    loaded_mlb = pickle.load(fp)
data_to_encode = df_country['RM'].apply(lambda x: [x] if not isinstance(x, list) else x)
encoded_data = loaded_mlb.transform(data_to_encode)
df_encoded = pd.DataFrame(encoded_data, columns=loaded_mlb.classes_)
df_country=df_country.join(df_encoded)
df_country.drop(columns=['RM'], inplace=True)
country_input = df_country.values
Input = pd.DataFrame(country_input , columns = loaded_model.feature_names)
Input_dmatrix = xgb.DMatrix(Input)

country_methane_predictions = loaded_model.predict(Input_dmatrix)
print(country_methane_predictions)
predictions=pd.DataFrame(country_methane_predictions)
predictions.to_csv('/content/Dairy_Grass_methane_predictions.csv', index=False)
```
