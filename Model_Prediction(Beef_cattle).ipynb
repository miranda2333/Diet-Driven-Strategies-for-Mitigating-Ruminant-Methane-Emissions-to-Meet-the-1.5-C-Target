```python
from catboost import CatBoostRegressor, Pool

df_country = pd.read_csv('/content/Beef_Mix_diet.csv', encoding='gbk')
df_country = df_country.iloc[:, 1:]

categorical_features = ['RM']

for feature in categorical_features:
    df_country[feature] = df_country[feature].fillna('').astype(str)

model = CatBoostRegressor()
model.load_model('/content/drive/MyDrive/TEST_Beef_Cat.cbm')
pool = Pool(data=df_country, cat_features=categorical_features)

predictions = model.predict(pool)
print(predictions)
predictions=pd.DataFrame(predictions)
predictions.to_csv('Beef_Grass_predictions.csv', index=False)
```
